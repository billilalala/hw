```python
#splitting the data

y_train = train_df['Gender']
x_train = train_df.drop(labels=['Gender'],axis=1)
x_train
```

```python
y_val = val_df['Gender']
x_val = val_df.drop(labels=['Gender'],axis=1)
x_val
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_val = scaler.transform(x_val)
```

```python
from keras.layers import Dropout

#####Specify the structure of a Neural Network

model = Sequential( ) # Model

model.add(Dense(10, activation='relu', input_shape=(11,), name = 'Input')) # Input Layer - need to speicfy the shape of inputs, 11 comes from the number of columns in the dataset
model.add(Dropout(0.5))
model.add(Dense(6, activation='relu', name='Hidden-Layer1')) # Hidden Layer1,
model.add(Dropout(0.5))
model.add(Dense(4, activation='relu', name='Hidden-Layer2')) # Hidden Layer2,
model.add(Dropout(0.5))
model.add(Dense(4, activation='relu', name='Hidden-Layer3')) # Hidden Layer3,
model.add(Dropout(0.5))
model.add(Dense(4, activation='relu', name='Hidden-Layer4')) # Hidden Layer4,
model.add(Dense(1, activation='sotmax', name='Output-Layer')) # Output Layer,


#model compilation

model.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics = ['Accuracy'], loss_weights = None, weighted_metrics= None)
```

```python
#fitting of the model on the dataset
#this is where we should add both the decay and the learning rate



miao = model.fit(x_train, y_train, batch_size = 100, epochs= 20, verbose = 2, callbacks = None, validation_split = 0.3, shuffle = True, class_weight=None, sample_weight = None, initial_epoch= 0, validation_data = (x_val, y_val) )
```

```python
pred_labels_tr = (model.predict(x_train) > 0.5).astype(int)
pred_labels_te = (model.predict(x_val) > 0.5).astype(int)


print("")
print('-------------------- Model Summary --------------------')
model.summary() # print model summary
print("")
print('-------------------- Layers --------------------')
for layer in model.layers:
    print("Layer: ", layer.name) # print layer name

print("")
print('---------- Evaluation on Training Data ----------')
print(classification_report(y_train, pred_labels_tr))
print("")

print('---------- Evaluation on Test Data ----------')
print(classification_report(y_val, pred_labels_te))
print("")
```
